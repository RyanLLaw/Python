## 2) Import pass and rushing stats. Clean Pass Stats. Be sure to check for duplicates. 20pts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

passing = pd.read_csv("pass_stats.csv")
rushing = pd.read_csv("run_stats.csv")

passing.head()
```

```python
cleanpass = passing.drop_duplicates()

cleanrun = rushing.drop_duplicates()
```

```python
cleanpass.isna().sum()
```

## 3) Merge the Data Sets 20pts


```python
# Hint, one key will not be enough. You could make your own or use a list of two.
stats = cleanpass.merge(cleanrun, on =["Player Code", "Game Code"])

stats.head()
```

## 4) Filter the data set so it only contains the player code 1008996  20pts

```python
cam = stats[stats["Player Code"]== 1008996]

cam.head()
```
